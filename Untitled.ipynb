```python
import pandas as pd
```

```python
df = pd.read_csv("Data/Omics-NORM_100count_1E6_cut_input_3.csv")
```

```python
df.head()
```

```python
microbe_list = ["Bacteria", "Archaea", "Eukaryota", "Organelle"]

microbe_df = df[df["Sample"].str.contains('|'.join(microbe_list))]
microbe_df = microbe_df.rename(columns={"Sample": "#OTU ID"})
metabolomics_df = df[~df["Sample"].str.contains('|'.join(microbe_list))]
metabolomics_df = metabolomics_df.rename(columns={"Sample": "#OTU ID"})

# Creating Taxonomy and Feature Metadata
microbe_taxonomy_df = pd.DataFrame()
microbe_taxonomy_df["featureid"] = microbe_df["#OTU ID"]
microbe_taxonomy_df["Taxonomy"] = microbe_df["#OTU ID"]
microbe_taxonomy_df.to_csv("microbe_metadata.tsv", sep="\t", index=False)

metabolomics_identifier_list = list(metabolomics_df["#OTU ID"])
metabolomics_metadata_list = []
for identifier in metabolomics_identifier_list:
    try:
        splits = identifier.split("_")
        feature_id = splits[0]
        mz = splits[1]
        rt = splits[2]
        ID = ":".join(splits[3:])
        
        output_dict = {}
        output_dict["sampleid"] = identifier
        output_dict["feature_id"] = feature_id
        output_dict["mz"] = mz
        output_dict["rt"] = rt
        output_dict["ID"] = ID
        
        metabolomics_metadata_list.append(output_dict)
        
    except KeyboardInterrupt:
        raise
    except:
        continue

pd.DataFrame(metabolomics_metadata_list).to_csv("metabolomics_metadata.tsv", sep="\t", index=False, columns=["sampleid", "feature_id", "mz", "rt", "ID"])
```

```python
microbe_df.to_csv("microbes.tsv", sep="\t", index=False)
metabolomics_df.to_csv("metabolomics.tsv", sep="\t", index=False)
!biom convert -i microbes.tsv -o microbes.biom --to-hdf5
!biom convert -i metabolomics.tsv -o metabolomics.biom --to-hdf5
```
